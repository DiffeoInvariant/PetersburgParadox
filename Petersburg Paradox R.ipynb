```r
PetersburgParadox <- function(samples, trials){
    avg_payoff = rep(0,samples)
    for(i in 1:samples){
        for(j in 1:trials){
            done  = FALSE
            count = 1
            while(!done){
                if(sample(c(0,1),1) == 0){
                    avg_payoff[j]  = avg_payoff[j] + 2^count
                    done = TRUE
                }
                else{
                    count = count +1
                }
            } #end while
        }
        avg_payoff[i] = avg_payoff[i]/trials
    }
    mu = mean(avg_payoff)
    std = sd(avg_payoff)
    plot(density(avg_payoff,kernel = "gaussian"))
}
```

```r
PetersburgParadox(1000,100)
```
